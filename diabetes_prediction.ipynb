**Feature Engineering**

```python
ages_count = [0, 25, 45, 65, df['age'].max()]
age_label = ['Young', 'Adult', 'MiddleAge', 'Senior']

df['age_bins'] = pd.cut(df['age'], bins=ages_count, labels=age_label, right=True)
```

```python
# Creating bins as per WHO standard

bmi_count = [0, 18.5, 24.9, 29.9, 34.9, df['bmi'].max()]
bmi_label = ['Underweight','Normal','Overweight','Obese','Extremely Obese']

df['bmi_bins'] = pd.cut(df['bmi'], bins=bmi_count, labels=bmi_label, right=True)
```

```python
# creating bins as per the clinical meaning

hbA1c_count = [0, 5.7, 6.4, df['hbA1c_level'].max()]
hbA1c_labels = ['Normal','Prediabetes','Diabetes']

df['hbA1c_level_bins'] = pd.cut(df['hbA1c_level'], bins=hbA1c_count, labels=hbA1c_labels, right=True)
```

```python
blood_glucose_count = [0, 100, 125, df['blood_glucose_level'].max()]
blood_glucose_labels = ['Normal','Prediabetes','Diabetes']

df['blood_glucose_level_bins'] = pd.cut(df['blood_glucose_level'], bins=blood_glucose_count, labels=blood_glucose_labels, right=True)
```

```python
def map_comorbidity(row):
    if row['hypertension'] == 0 and row['heart_disease'] == 0:
        return "No Comorbidity"
    elif row['hypertension'] == 1 and row['heart_disease'] == 0:
        return "Hypertension"
    elif row['hypertension'] == 0 and row['heart_disease'] == 1:
        return "Heart Disease"
    else:
        return "Hypertension + Heart Disease"

df['comorbidities'] = df.apply(map_comorbidity, axis=1)
```

```python
top_10_states = df['location'].value_counts().nlargest(10).index

df['location_group'] = df['location'].apply(lambda x: x if x in top_10_states else 'other')
```

**Encoding**

```python
df_encoded = pd.get_dummies(
    data=df,
    columns=['year','gender','smoking_history','age_bins','bmi_bins',
             'hbA1c_level_bins','blood_glucose_level_bins',
             'comorbidities','location_group'],
    drop_first=True
)
```

```python
numerical_cols = df_encoded.select_dtypes(include = ['bool','uint8','int64']).columns
df_encoded[numerical_cols] = df_encoded[numerical_cols].astype(int)
```

```python
import re
import string
import nltk
from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
nltk.download('averaged_perceptron_tagger_eng')
#  Setup for Colab (download both punkt + punkt_tab)
for pkg in ["punkt", "punkt_tab", "stopwords", "wordnet", "omw-1.4"]:
    try:
        nltk.data.find(f"corpora/{pkg}" if "punkt" not in pkg else f"tokenizers/{pkg}")
    except LookupError:
        nltk.download(pkg)

#  Stopwords
stopwords_set = set(stopwords.words('english'))
negations = {"no", "not", "nor", "never", "without"}
stopwords_set -= negations

#  Lemmatizer
lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(word):
    """Get POS tag for accurate lemmatization"""
    tag = pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

def preprocess(text):
    """Lowercase, remove punctuation, tokenize, lemmatize, remove stopwords"""
    text = str(text).lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = word_tokenize(text)
    tokens = [lemmatizer.lemmatize(word, get_wordnet_pos(word))
              for word in tokens if word not in stopwords_set]
    return ' '.join(tokens).strip()

#  Apply preprocessing
df_encoded['clinical_notes'] = df_encoded['clinical_notes'].apply(preprocess)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from scipy.sparse import hstack
from imblearn.over_sampling import SMOTE

X_numeric = df_encoded.drop(columns=['diabetes', 'clinical_notes'])
text_data = df_encoded['clinical_notes']
y = df_encoded['diabetes']

X_train_num, X_test_num, train_notes, test_notes, y_train, y_test = train_test_split(
    X_numeric, text_data, y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

scaler = MinMaxScaler()
X_train_num_scaled = scaler.fit_transform(X_train_num)
X_test_num_scaled = scaler.transform(X_test_num)

tfidf = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
X_train_notes_tfidf = tfidf.fit_transform(train_notes)
X_test_notes_tfidf = tfidf.transform(test_notes)

X_train_final = hstack([X_train_num_scaled, X_train_notes_tfidf])
X_test_final = hstack([X_test_num_scaled, X_test_notes_tfidf])

smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_final, y_train)

print("Before SMOTE:", X_train_final.shape, y_train.value_counts())
print("After SMOTE:", X_train_balanced.shape, y_train_balanced.value_counts())
```
